```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame({'whoAmI':lst})
df.head()
```

```python
lst2 = []
for i in lst:
  record = {'human':0, 'robot':0}
  if i == 'robot':
    record['robot'] = 1
  else:
    record['human'] = 1
  lst2.append(record)
df2 = pd.DataFrame(lst2)
df2.head()
```
